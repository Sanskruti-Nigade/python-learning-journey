# Titanic Dataset Analysis
Cleaning, summary statistics, and visual exploration of the Titanic dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

## Summary Statistics

```python
df.describe(include='all')
```

## Data Cleaning
- Fill missing `Age` with the median
- Fill missing `Embarked` with the mode
- Drop `Cabin` (mostly missing)

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns=['Cabin'])
df.isnull().sum()
```

## Plot 1: Passenger Survival Count

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Survived', hue='Survived', palette='Set2', legend=False)
plt.title('Passenger Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.show()
```

## Plot 2: Survival Rate by Passenger Class

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Pclass', y='Survived', hue='Pclass', palette='viridis', legend=False, errorbar=None)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

## Plot 3: Age Distribution

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='Age', bins=30, kde=True, color='steelblue')
plt.title('Passenger Age Distribution')
plt.show()
```

## Plot 4: Correlation Heatmap

```python
plt.figure(figsize=(7, 5))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Numeric Columns)')
plt.show()
```
